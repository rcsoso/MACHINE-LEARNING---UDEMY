## La simulación de Monte Carlo

1. Generamos dos números aleatorios uniforme (x e y) entre 0 y 1.
2. Calcularemos x.*x + y.*y.
* Si el valor es inferior a 1 --> estamos dentro del círculo.
* Si el valor es superior a 1 --> estamos fuera del círculo.
3. Calculamos el número de total de veces que están dentro del círculo y lo dividimos entre el número total de intentos para obtener una aproximación de la probabilidad de caer dentro del círculo.
4. Usamos dicha probabilidad para aproximar el valor de π.
5. Repetimos el experimento un número suficiente de veces , para obtener  diferentes aproximaciones de π.
6. Calculamos el promedio de los experimentos anteriores para dar un valor final de π.

```python
def pi_montecarlo(n,n_exp): # n =  Número de puntos por experimento n_exp = Número de experimentos
    lista_pi = []
    for i in range(n_exp):
        valor = 0
        x = np.random.uniform(0,1,n)
        y = np.random.uniform(0,1,n)
        z = x**2 + y**2
        for _ in z:
            if _ <= 1:
                valor += 1
        lista_pi.append(valor * 4 / n)
    pi =  np.mean(lista_pi) 
    return (pi,lista_pi)
```

```python
pi,lista_pi= pi_montecarlo(10000,200)
print(pi)
plt.plot(lista_pi)
```
